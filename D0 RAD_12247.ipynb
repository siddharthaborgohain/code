```python
def value_interpreter(number,limit):
    import math
    final_val = math.ceil(number/limit)
    list_values = [0]
    for i in range(final_val):
        value_append = (i+1)*limit
        if value_append<number:
            list_values.append((i+1)*limit)
        else:
            list_values.append(number)
    return final_val,list_values
```

```python
def query_run_with_batch_size(host,port,user,password,db_name,batch_size,query_start,query_end,awb_to_find):
    """
    Either pass as List or as Dataframe for awb_to_find
    """
    #import pandas as pd
    import pymysql
    import warnings
    warnings.filterwarnings("ignore")
    #looping over
    data_list_1 = []

    if type(awb_to_find) != pd.core.frame.DataFrame:
        y = pd.DataFrame({'val_to_find':awb_to_find})
    else:
        y = awb_to_find
        
    a,b = value_interpreter(len(y[y.columns[0]]),batch_size)
    for i,j in enumerate(b):
        if ((i+1)==len(b)):
            pass
        else:
            print('Working Range : ',str(j),'-',str(b[i+1]))

            query_data = query_start+ (',').join(y[y.columns[0]][j:b[i+1]].apply(lambda x:str(int(x)))) +query_end
            conn = pymysql.connect(host= host , port=port , user=user , passwd=password, db=db_name)
            print(conn)
            dt1 = datetime.datetime.now()
            temp_data = pd.read_sql(query_data,conn)
            dt2 = datetime.datetime.now()
            print('time takent to execute and fetch the query: ' ,dt2-dt1)
            print(temp_data.shape)
            print('Saving_data of Batch: %d of Total Batch %d\n'%(i+1,len(b)-1))
            data_list_1.append(temp_data)
    
    data = pd.concat(data_list_1)
    print(data.shape)
    return data
```
